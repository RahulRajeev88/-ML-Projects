```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/agroData.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df=df.drop(['Grade'],axis=1)
df.head()
```

```python
print('The Average values of the minimum price:',df['Min_x0020_Price'].mean())
```

```python
print('The Average values of the maximum price:',df['Max_x0020_Price'].mean())
```

```python
#fill the null values present in minimum and maximum value by its mean value or average value
#fillna() is a method that replaces missing (NaN) values with the specified value.
#The inplace=True parameter ensures that the changes are made directly to the DataFrame 'df' without creating a new copy.
df['Min_x0020_Price'].fillna(df['Min_x0020_Price'].mean(),inplace=True)
```

```python
df['Max_x0020_Price'].fillna(df['Max_x0020_Price'].mean(),inplace=True)
```

```python
#Checking Duplicate values in the given data
print('The duplicate values in the dataframe is:',df.duplicated().sum())
```

```python
#Describe the statistics in the data
df.describe()
```

```python
#Plotting correlation map using the heatmap
sns.heatmap(df.corr(),annot=True,cmap='BrBG',fmt='.2f',linewidths=1)
plt.title('The Correrlation Map')
plt.show()
```

**EDA:EXPLORATORY DATA ANALYSIS.**

**To find the unique states in the data, you can use the unique() function on the 'State' column:**

```python
unique_states=df['State'].unique()
print('List of states present in the data:',unique_states)
```

```python
#Highest dominated state
dominated_state=df['State'].value_counts().idxmax()
print('Highest dominated_state in the data:',dominated_state)
```

```python
#Least dominated state
Least_state=df['State'].value_counts().idxmin()
print('Least dominated_state in the data:',Least_state)
```

```python
max_and_min_price=(df['Min_x0020_Price'].min(),df['Min_x0020_Price'].max())
print('The Maximum Value and The Minimum Value lies between in the range of:',max_and_min_price)
```

```python
most_commodity=df['Commodity'].value_counts().idxmax()
print('The most demanding commodity in the data is:',most_commodity)
```

```python
least_commodity=df['Commodity'].value_counts().idxmin()
print('The least demanding commodity in the data is:',least_commodity)
```

```python
#create Bar chart for states
df['State'].value_counts().sort_values(ascending=False)\
.plot(kind='bar',title='States in India',figsize=(8,6),color=['#164B60','#1B6B93','#4FC0D0','#A2FF86','#4C4B16'])
plt.xlabel('States')
plt.ylabel('Value counts')
plt.show()
```

```python
#Top 10 Districts using barchart
df['District'].value_counts().nlargest(10).sort_values(ascending=False)\
.plot(kind='bar',title='Top 10 Districts in India',figsize=(8,6),color=['#0080FF','#E1F5A9','#01DF01','#F7FE2E'])
plt.xlabel('District')
plt.ylabel('value_count')
plt.show()
```

```python
# Calculate the most revenue generate in the market
df['Market_revenue']=df['Min_x0020_Price']+df['Max_x0020_Price']+df['Modal_x0020_Price']
label=['Nahan','Kollengode','Mumbai','Thodupuzha','Fish,Poultry & Egg Market, Gazipur','Pothencode','Shimoga','Yellapur','Siddapur','Palakkad']
df.groupby('Market')['Market_revenue'].sum().sort_values(ascending=False).head(10)\
.plot(kind='pie',labels=label,
    colors=['#81F7D8','#BCA9F5','#2EFE64','#FE2E2E','#FF0000'],
    autopct='%1.2f%%',
    shadow=True)
plt.title("Most Revenued to Market in the data")
plt.show()
```

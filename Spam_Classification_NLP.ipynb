```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/SPAM text message 20170820 - Data.csv')
df
```

**Data Cleaning & Text Preprocessing**

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
stopwords_list=[]
for word in stopwords.words('english'):
  stopwords_list.append(word)
```

```python
stopwords_list
```

```python
from nltk.stem.porter import PorterStemmer
porter=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(df)):
  review=re.sub('[^a-zA-Z0-9]',' ',df['Message'][i])
  review=review.lower()
  review=review.split()
  review=[porter.stem(word) for word in review if not word in stopwords_list]
  review=' '.join(review)
  corpus.append(review)
```

```python
corpus
```

**Train Test Split**

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vec=CountVectorizer(max_features=2500)
```

```python
X=count_vec.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
df['Category']
```

```python
y=pd.get_dummies(df['Category'],drop_first=True)
y

```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
count_vec.vocabulary_
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB
model1=GaussianNB()
model1.fit(X_train,y_train)
```

```python
y_pred=model1.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
model2=MultinomialNB()
model2.fit(X_train,y_train)
y_pred=model2.predict(X_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)*100
```
